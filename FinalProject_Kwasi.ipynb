```python
#define funciton to convert fasta to Nexus file
def convertFastaToNexus():
    #import phylo and SeqIO
    from Bio import Phylo, SeqIO
    #define list of regions
    filenames = ['northEast','southEast','midWest','mountain','southWest','pacific', "NoMask", "HighestVax", "limitedOutdoor", "lowestPartial"]
    #iter through filenames
    for i in filenames:
        #define file path 
        file = f"./Alignment/{i}_alignment.fasta"
        out = f"./Alignment/{i}_alignment.nexus"
        #convert MSA fasta to nexus format
        SeqIO.convert(file, "fasta", out,"nexus","protein")
    return
```

```python
def generatePhylo():
    #import phylo and AlignIO
    from Bio import Phylo, AlignIO
    #import Distance Calculator and DistanceTreeConstructor
    from Bio.Phylo.TreeConstruction import DistanceCalculator, DistanceTreeConstructor
    #define filenames list
    filenames = ['northEast','southEast','midWest','mountain','southWest','pacific', "NoMask", "HighestVax", "limitedOutdoor", "lowestPartial"]
    #iter through filenames
    for i in filenames[7:8]:
        #define file path
        file = f"./Alignment/{i}_alignment.nexus"
        #create MSA alignment object
        aln = AlignIO.read(file, 'nexus')
        #define calculator 
        calc = DistanceCalculator('identity')
        #calculate distnace measurement
        dm = calc.get_distance(aln)
        #define constructor
        constructor = DistanceTreeConstructor(calc, 'nj')
        trees = constructor.build_tree(aln)
        #generate phylogenetic tree
        #create path to write tree files
        out = f"./Trees/{i}_phylo.xml"
        #write trees to new xml file
        Phylo.write(trees, out, 'phyloxml')

    return
    
```

```python
#write function to draw trees
def drawTrees():
    #import phylo
    from Bio import Phylo
    import matplotlib
    import matplotlib.pyplot as plt
    #from Phylo import Phylo.PhyloXMLIO
    #define file names
    filenames = ['northEast','southEast','midWest','mountain','southWest','pacific', "NoMask", "HighestVax", "limitedOutdoor", "lowestPartial"]
    #iter through filenames
    for i in filenames:
        #define file path
        file = f"./Trees/{i}_phylo.xml"
        #read each tree
        #trees = Phylo.PhyloXMLIO.parse(file)
        trees = Phylo.read(file,'phyloxml')
        #draw trees
        #trees.ladderize()
        #fig = plt.figure(figsize=(20,20), dpi=100)
        #axes = fig.add_subplot(1,1,1)  
        fig = plt.figure(figsize=(50,500), dpi=100) 
        matplotlib.rc('font', size=12)              
        matplotlib.rc('xtick', labelsize=10)       
        matplotlib.rc('ytick', labelsize=10)       
        #turtle_tree.ladderize()		    
        axes = fig.add_subplot(1, 1, 1)
        Phylo.draw(trees, axes=axes, branch_labels=lambda x: x.name)
        #print(trees)
        fig.savefig(f"./Trees/{i}")
        
        #plt.show()
    return
```
